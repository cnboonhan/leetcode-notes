# Two Pointers

## Pair with Target Sum


```python
def f(A, S):
    # A: int[] | sorted(A)
    # S: int 
    # x: (i, j) | A[i] + A{j] = S
    # i, j | i, j < len(A)
    i = 0
    j = len(A) - 1
    while (i<j):
        current_sum = A[i] + A[j]
        if current_sum == S:
            return [i, j]
        if current_sum > S:
            j -= 1
            continue
        if current_sum < S:
            i += 1
            continue
    raise Exception
    
# O(N) time, O(1) space
assert f([1, 2, 3, 4, 6], 6) == [1, 3]
assert f([2, 5, 9, 11], 11) == [0, 2]
```

## Remove Duplicates


```python
import math
def f(A):
    # A: int[] | sorted(A)
    # x: int | x = len(uniq(A))
    l = 0
    prev = math.inf
    for i in range(len(A)):
        # invariant: l contains count of unique elements of A
        if prev == A[i]:
            continue
        else:
            l += 1
            prev = A[i]
    return l

assert f([2, 3, 3, 3, 6, 9, 9]) == 4
assert f([2, 2, 2, 11]) == 2
```

```python
def f(A, key):
    # A: int[] | unsorted(A)
    # key: int 
    # x : int | len(B)
    # B : int[] | B contains all of A except key
    B_end = 0
    for i in range(len(A)):
        # invariant: subarray [0, B_end] contains all non-key elements of A
        if A[i] == key:
            continue
        else:
            A[B_end] = A[i]
            B_end += 1
    return B_end

assert(f([3, 2, 3, 6, 3, 10, 9, 3], 3) == 4)
assert(f([2, 11, 2, 2, 1], 2) == 2)
```

## Squaring a Sorted array


```python
def f(A):
    # A: int[] | sorted(A)
    # X: int[] | sorted(X), eX contains squared values of all eA
    i = 0
    j = len(A) - 1
    Y = []
    while i < j:
        # invariant, Y contains elements of A in reverse sorted order
        A_i_sq = A[i] ** 2
        A_j_sq = A[j] ** 2
        if A_i_sq > A_j_sq:
            Y.append(A_i_sq)
            Y.append(A_j_sq)
        else:
            Y.append(A_j_sq)
            Y.append(A_i_sq)
        i+=1
        j-=1
    if i == j:
        Y.append(A[i] ** 2)
    Y.reverse()
    return Y

assert(f([-2, -1, 0, 2, 3]) == [0, 1, 4, 4, 9])
assert(f([-3, -1, 0, 1, 2]) == [0, 1, 1, 4, 9])
```
