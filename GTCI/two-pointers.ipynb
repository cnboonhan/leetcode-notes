# Two Pointers

## Pair with Target Sum


```python
def f(A, S):
    # A: int[] | sorted(A)
    # S: int 
    # x: (i, j) | A[i] + A{j] = S
    # i, j | i, j < len(A)
    i = 0
    j = len(A) - 1
    while (i<j):
        current_sum = A[i] + A[j]
        if current_sum == S:
            return [i, j]
        if current_sum > S:
            j -= 1
            continue
        if current_sum < S:
            i += 1
            continue
    raise Exception
    
# O(N) time, O(1) space
assert f([1, 2, 3, 4, 6], 6) == [1, 3]
assert f([2, 5, 9, 11], 11) == [0, 2]
```
