# Hash Tables

## Pair with Target Sum


```python
def f(A, S):
    # A: int[] | sorted(A)
    # S: int 
    # x: (i, j) | A[i] + A{j] = S
    # i, j | i, j < len(A)
    
    D = {}
    for i in range(len(A)):
        target_Ai = S - A[i]
        if target_Ai in D.keys():
            return [D[target_Ai], i]
        else:
            D[A[i]] = i
            # invariant: D is hashtable with keys containing seen values of A. 
            
    raise Exception
    
# O(N) time, O(N) space
assert f([1, 2, 3, 4, 6], 6) == [1, 3]
assert f([2, 5, 9, 11], 11) == [0, 2]
```
